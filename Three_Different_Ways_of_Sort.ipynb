# Three_Different_Ways_of_Sort

# sort method를 이용한 소트

sort method는 inplace function이다.

```python
import numpy as np

np.random.seed(123456789)

x = np.random.uniform(0., 1., (5,))
print(x)

if 1:
    # sort method는 inplace function이다.
    x.sort() # inplace function
elif 1:
    # sort method는 inplace function이다. 
    # return 값이 없다. 
    # 따라서 default로 None이 리턴된다.
    x = x.sort()

print(x)
```

# np.sort를 이용한 소트

```python
import numpy as np

np.random.seed(123456789)

x = np.random.uniform(0., 1., (5,))
x_sorted = np.sort(x)

print(x)
print(x_sorted)
```

# np.argsort를 이용한 소트

```python
import numpy as np

np.random.seed(123456789)

x = np.random.uniform(0., 1., (5,))
id_x = np.argsort(x)

print(x)
print(id_x)
print(x[id_x])
print(x[np.argsort(x)])
```

# $y=x^2$ 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-1., 1., (100,))
y = x**2

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-1., 1., (100,))
y = x**2

plt.plot(x, y, 'o')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-1., 1., (100,))
if 0:
    x.sort()
elif 0:
    x = np.sort(x)
elif 1:
    x = x[np.argsort(x)]
    
y = x**2

plt.plot(x, y)
plt.show()
```
